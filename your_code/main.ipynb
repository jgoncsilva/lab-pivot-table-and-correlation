#### Is there a correlation between study time and Grade? Use both Pearson and Spearman correlation and comment your results.

```python
# your answer here
p_corr = time_grades['Grade'].corr(time_grades['Study time'], method='pearson')
sp_corr = time_grades['Grade'].corr(time_grades['Study time'], method='spearman')

print(f'Pearson correlation : {p_corr})\nSpearman correlation : {sp_corr}')
```
